## Dealing with Null Values

```python
print('column name' + '\t\t\t' + 'IsUnique')
df.apply(lambda x: x.is_unique,axis=0)
```

```python
# we shall drop id since its unique 
df = df.drop('id', axis=1)
```

```python
df.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
df.shape
# (row x columns)
```

```python
df_dropna_row = df.dropna()
print("Dataframe shape after dropping only rows with null value", df_dropna_row.shape)

# Drop all columns that contain null value
df_dropna_col = df.dropna(axis=1)
print("Dataframe shape after dropping only columns with null value", df_dropna_col.shape)
```
