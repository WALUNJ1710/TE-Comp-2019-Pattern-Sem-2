**Provide summary statistics (mean, median, minimum, maximum, standard deviation) for 
a dataset (age, income etc.) with numeric variables grouped by one of the qualitative 
(categorical) variable. **

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df
```

```python
df.describe()
```

```python
df['Gender'].value_counts()
```

```python
gr = df.groupby('Gender')
```

```python
gr.get_group('Male')
```

```python
gr.get_group('Male').describe()
```

```python
gr_age = df.groupby('Age')
```

```python
gr_age.get_group(27)
```

```python
gr.agg(np.mean)
```

```python
gr.agg(np.max)
```

```python
gr.agg(np.min)
```

```python
gr.agg(np.std)
```

```python
gr.agg(np.median)
```

```python
gr_age.agg(np.mean)
```

```python
gr_age.agg(np.max)
```

```python
gr_age.agg(np.min)
```

```python
gr_age.agg(np.std)
```

```python
gr_age.agg(np.median)
```
