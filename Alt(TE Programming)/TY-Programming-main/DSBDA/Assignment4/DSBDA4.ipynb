```python
import pandas as pd
import numpy as np
```

```python
df=  pd.read_csv('BostonHousing.csv')
```

```python
df
```

**Separate input and output**

```python
x = df.drop('medv', axis =1)
y = df['medv']
```

```python
x.shape
```

```python
y
```

```python
df.corr()       #show correlation
```

```python
df.describe()
```

**Feature SCALING**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
pd.DataFrame(x_scaled).describe()
```

**Build a model by Linear Regression**

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_scaled, y )
```

**In a Jupyter environment, please rerun this cell to show the HTML representation or trust the notebook.
On GitHub, the HTML representation is unable to render, please try loading this page with nbviewer.org.**

```python
### With the help of r2 score we can find linearity of the model
```

```python
regressor.score(x_scaled,y)
```

```python
### So 74% linearnity is show by this method. so 74% equivalent accuracy .
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y,regressor.predict(x_scaled))
#this is coefficient of determination
```

```python
### predict the value of prices of the house using the given features
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y,regressor.predict(x_scaled))
```
