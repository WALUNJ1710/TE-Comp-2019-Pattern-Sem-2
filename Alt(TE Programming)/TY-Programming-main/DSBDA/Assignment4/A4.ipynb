```python
df['CRIM'].fillna(method='ffill',inplace=True)
df['ZN'].fillna(method='ffill',inplace=True)
df['INDUS'].fillna(method='ffill',inplace=True)
df['CHAS'].fillna(method='ffill',inplace=True)
df['AGE'].fillna(method='ffill',inplace=True)
df['LSTAT'].fillna(method='ffill',inplace=True)
```

```python
df.isnull().sum()
```

**Define independant and dependant variable from dataset**

```python
x=df.iloc[:,0:13]
y=df.iloc[:,-1]
```

**Splitting data into training and testing dataset**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

**Shaping training and testing data**

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

**Standardizing the data**

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

**Importing Linear regression model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,Y_train)
```

```python
pred=lr.predict(X_test)
print(pred)
```

**Root means sqaure root**

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(Y_test,pred))
```

```python
print('Root means squared value is: ',rmse)
```

**Training accuracy**

```python
print('Training Accuracy is: ',lr.score(X_train,Y_train))
```

**Testing Accuracy**

```python
print("Testing Accuracy is: ",lr.score(X_test,Y_test))
```
