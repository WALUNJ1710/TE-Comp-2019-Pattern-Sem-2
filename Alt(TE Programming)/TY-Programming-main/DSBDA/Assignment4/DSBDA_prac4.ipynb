**Feature Description**

1) **CRIM** - per capita crime rate by town

2) **ZN** - proportion of residential land zoned for lots over 25,000 sq.ft.

3) **INDUS** - proportion of non-retail business acres per town

4) **CHAS** - Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)

5)**NOX** - nitric oxides concentration (parts per 10 million)

6) **RM** - average number of rooms per dwelling

7) **AGE** - proportion of owner-occupied units built prior to 1940

8) **DIS** - weighted distances to five Boston employment centres

9) **RAD** - index of accessibility to radial highways

10) **TAX** - full-value property-tax rate per '10,000 Dollar'

11) **PTRATIO** - pupil-teacher ratio by town

12) **B** - 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town

13) **LSTAT** - lower status of the population

14) **MEDV** - Median value of owner-occupied homes in 1000's Dollar)

**Correlation between Feature**

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True);
```

**Detecting outlier**

```python
# Create a figure and a set of subplots with as many rows as needed and 3 columns

#calculate number of row needed
num_rows = (len(df.columns) + 2)//3
fig,axes = plt.subplots(nrows = num_rows, ncols=3, figsize=(15,15))

#creating boxplots for each column in DataFrame
for i, col in enumerate(df.columns):
  sns.boxplot(data=df, x=col, ax=axes[i // 3, i % 3])
```

**Data Visualisation**

```python
fig = plt.figure(figsize=(15,5))
fig.add_subplot(1,2,1)
sns.countplot(x=df['CHAS']);
fig.add_subplot(1,2,2)
sns.countplot(x=df['RAD']);
```
